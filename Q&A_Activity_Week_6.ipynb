### 3. Find out how many countries with respect to the region are available in the dataset.

```python
data.groupby('Region')[['country']].count()
```

```python
reg = data['Region'].value_counts()
plt.figure(figsize=(16,8))
sns.barplot(x=reg.index,y=reg.values,palette='Set1')
plt.title('Number of Countries by Region',fontweight='bold',fontsize=15)
plt.xlabel('Region',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=90)
plt.yticks(np.arange(0,22,2))
plt.grid()
plt.show()
```

### 4 a). On average which region has the highest GDP per capita (current US$)


```python
GDP_high=data.groupby("Region")[["GDP per capita (current US$)"]].mean()
GDP_high.sort_values(by="GDP per capita (current US$)",ascending=False)
GDP_high.nlargest(1,"GDP per capita (current US$)")
```

### 4 b). On average which region has the highest International trade: Imports (million US$)

```python
ITI_high=data.groupby("Region")[["International trade: Imports (million US$)"]].mean()
ITI_high.sort_values(by="International trade: Imports (million US$)",ascending=False)
ITI_high.nlargest(1,"International trade: Imports (million US$)")
```

### 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
GDP_cons=data.groupby("Region")[["GDP per capita (current US$)"]].std()
GDP_cons.sort_values(by="GDP per capita (current US$)")
```

```python
GDP_cons.nsmallest(1,"GDP per capita (current US$)")
```
