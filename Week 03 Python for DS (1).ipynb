### 4.Read the given data set and answer the following questions. Show the steps taken to produce your final answer.

1. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
2. How many patients in the data are having the blood pressure level above 120?
3. Select the rows the age is between 24 and 29 (inclusive)
4.Find the minimum and maximum value of insulin level in the data.
5. Create a histogram of "Age".
6. Create a scatter plot of "BMI" vs "Glucose".
7. Replace the min value of "BMI" with mean value of BMI
8. Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

```python
diabetes_data= pd.read_csv("diabetes_model.csv")
diabetes_data
```

```python
#1
diabetes_data[['Glucose', 'Age']][45:61]
```

```python
#2

df = diabetes_data[diabetes_data["BloodPressure"] > 120]
print('No of patients having the blood pressure level above 120 is',df.shape[0])
df
```

```python
#3

diabetes_data[(diabetes_data ['Age']>=24) & (diabetes_data ['Age']<= 29)]
```

```python
#4

min_value= diabetes_data['Insulin'].min()
max_value= diabetes_data['Insulin'].max()
print("Minimum Value of Insulin level in the data :",min_value)
print("Maximum Value of Insulin level in the data :",max_value)
```

```python
#5

plt.figure(figsize=(8,8))
plt.hist(diabetes_data['Age'],color="Green",rwidth=0.9)
plt.title('Histogram of Age',fontsize=22)
plt.xlabel("Age",fontsize=13)
plt.ylabel("Frequency",fontsize=13)
plt.xticks()
plt.yticks()
```

```python
#6

plt.figure(figsize=(8,6))
plt.scatter(diabetes_data['BMI'],diabetes_data['Glucose'],s=12,c ='red', marker='^')
plt.title('Plot of BMI  vs Glucose', fontsize = 16)
plt.xlabel('BMI')
plt.ylabel('Glucose')
```

```python
#7 a)

bmi_min = diabetes_data['BMI'].min()
bmi_mean = diabetes_data['BMI'].mean()

print('The min value of BMI is',bmi_min)
print('The mean value of BMI is', bmi_mean)
```

```python
#7 b)
diabetes_data.loc[(diabetes_data["BMI"]==0),"BMI"]=diabetes_data["BMI"].mean()
diabetes_data['BMI'].min()
```

```python
#8

plt.figure(figsize=(8,6))
sns.boxplot(x='Outcome',y='Glucose',data=diabetes_data)
plt.title("Boxplot of Outcome vs Glucose ", fontsize=20)
```

### 5.

**-** Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

**-** Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

**-** Combine the arrays first and second vertically, the final array size should be 10 x 5

```python
arr_n1 = np.zeros((5,5))
print("Array 1 is")
print(arr_n1)

arr_n2 = np.ones((5,5))
print("\n")
print("Array 2 is")
print(arr_n2)

arr_n = np.concatenate((arr_n1,arr_n2),axis=0)
print("\n")
print("Combining Array 1 & Array 2 is")
print(arr_n)
```
