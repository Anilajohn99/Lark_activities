```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
df
```

### 1. Compare churn count with respect to gender.

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='gender',hue='Churn')
plt.title('Churn Count VS Gender',fontweight='bold')
plt.show()
```

### 2. Find out how many female senior citizens there in the dataset.

```python
fsc = df[(df['gender']=='Female') & (df['SeniorCitizen']==1)]
fsc
```

```python
print('Number of female senior citizens in the datset is ',fsc["gender"].count())

plt.figure(figsize=(4,6))
sns.countplot(x=fsc['gender'])
plt.title('Count of female senior citizens', fontweight='bold', fontsize=16)
plt.show()
```

### 3. Compare 'tenure' with 'Total Charges'


```python
plt.figure(figsize=(8,6))
plt.scatter(df['tenure'],df['TotalCharges'],s=8)
plt.xlabel('Tenure', fontsize =12)
plt.ylabel('Total Charges', fontsize =12)
plt.title('Tenure Vs Total Charges', fontweight='bold',fontsize=16)
plt.show()
```

### 4. Find out which contract is preferred by the senior citizen.

```python
sc=df.loc[(df['SeniorCitizen']==1)]
sc
```

```python
print('Contract Preferences of Senior Citizens : \n',sc['Contract'].value_counts())

plt.figure(figsize=(8,6))
sns.countplot(x=sc['Contract'])
plt.title('Contract Preferences of Senior Citizens', fontweight='bold', fontsize=16)
plt.xlabel('Contract',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.grid()
plt.show()
```

### 5. Comment on your finds on Payment Method.

```python
print('Payment Method: \n',df['PaymentMethod'].value_counts())

plt.figure(figsize=(10,6))
sns.countplot(x=df['PaymentMethod'])
plt.title('Payment Method', fontweight='bold', fontsize=16)
plt.xlabel('Payment Type',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=10)
plt.grid()
plt.show()
```

```python
#genderwise payment method.
plt.figure(figsize=(10,6))
sns.countplot(x=df['PaymentMethod'], hue=df['gender'])
plt.title('Payment Method-Genderwise', fontweight='bold', fontsize=16)
plt.xlabel('Payment Type',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=10)
plt.grid()
plt.show()
```

```python
#Payment method among senior citizen.

plt.figure(figsize=(10,6))
sns.countplot(x=sc['PaymentMethod'])
plt.title('Payment Method among Senior Citizen', fontweight='bold', fontsize=16)
plt.xlabel('Payment Type',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=10)
plt.grid()
plt.show()
```

```python
# Contractwise Payment Method
plt.figure(figsize=(10,6))
sns.countplot(x=df['PaymentMethod'], hue=df['Contract'])
plt.title('Payment Method- Contractwise',fontweight='bold',fontsize=16)
plt.xlabel('Payment Type',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=10)
plt.grid()
plt.show()
```

```python
# Churn wise Payment Method
plt.figure(figsize=(10,6))
sns.countplot(x=df['PaymentMethod'],hue=df['Churn'])
plt.title('Payment Method- Churnwise',fontweight='bold',fontsize=16)
plt.xlabel('Payment Type',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=10)
plt.grid()
plt.show()
```
